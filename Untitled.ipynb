```python
import os
import django
os.environ.setdefault('DJANGO_SETTINGS_MODULE', 'SamparkSetu.settings')
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
django.setup()
    
import csv

from Mandal.models import MandalProfile
from Yuvak.models import YuvakProfile
yuvakmandal = MandalProfile.objects.filter(Name="Utsav").first()
print(yuvakmandal)
```

```python
with open('all yuvak.csv') as csv_file:
    reader = csv.reader(csv_file)
    # header = next(reader)
    ct = 0
    for row in reader:
        if not YuvakProfile.objects.filter(WhatsappNo=row[3]).exists():
            ct += 1
            yuvak = YuvakProfile(FirstName=row[0], 
                                 MiddleName=row[1], 
                                 SurName=row[2], 
                                 WhatsappNo=row[3],
                                 HomePhoneNo=row[4],
                                 Soc_Name=row[5],
                                 LandMark=row[6],
                                 Area=row[7],
                                 mandal_id=yuvakmandal.pk)
            yuvak.save()
            print(ct, row[0], row[2])
```
